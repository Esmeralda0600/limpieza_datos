```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
torre = ("../data/001_raw/002_Torre2.csv")
may = ("../data/002_intermediate/29-may-2026-limpio.parquet")
df_torre = pd.read_csv(torre,
                    index_col=0,
                    parse_dates=True
                    #date_format="%d%m%Y"
            )
df_may = pd.read_parquet(may)

#df_may.index
df_torre.index

```

```python
#concatenar 002_Torre2.csv y 29-mayo-2026.csv 
df_torre.index = pd.to_datetime(df_torre.index, format="mixed").tz_localize(None)
torre

torre_min = df_torre.copy()
may_min = df_may.copy()

torre_min.index = torre_min.index.floor("min")
may_min.index = may_min.index.floor("min")

torre_min = torre_min.groupby(level=0).mean()

concatenar= pd.concat(
    [torre_min, may_min],
    axis=1,
    join="inner"
)

concatenar


# df_torre.index
```

```python
concatenar.to_parquet("../data/003_processed/torre_29may.parquet")
```
