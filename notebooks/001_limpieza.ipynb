```python
import pandas as pd
import matplotlib.pyplot as pyplot
```

```python
may = ("../data/001_raw/29-may-2026.csv")
df_may = pd.read_csv(may,
                     encoding="ISO-8859-1",
                     index_col=11,
                     #parse_dates=True,

                     #date_format="%d%m%Y"
                     )

df_may.head()
df_may.index

```

```python
df_may.index = pd.to_datetime(df_may.index, format="mixed")
```

```python
#limpiar archivo sin columnas vacias en formato parquet

df_may = df_may.rename (columns={
    "Temperatura":"Temp",
    "Bulbo húmedo": "BH",
    "Punto de condensación": "PC",
    "% HR": "HR"
    })


df_may["Temp"] = df_may["Temp"].replace("°C", "", regex=True)
df_may["BH"] = df_may["BH"].replace("°C", "", regex=True)
df_may["PC"] = df_may["PC"].replace("°C", "", regex=True)
df_may["HR"] = df_may["HR"].replace("%", "", regex=True)
df_may["CO2"] = df_may["CO2"].replace("ppm", "", regex=True)


df_may["Temp"] = pd.to_numeric(df_may["Temp"])
df_may["BH"] = pd.to_numeric(df_may["BH"])
df_may["PC"] = pd.to_numeric(df_may["PC"])
df_may["HR"] = pd.to_numeric(df_may["HR"])
df_may["CO2"] = pd.to_numeric(df_may["CO2"])

df_may = df_may[["Temp", "BH", "PC", "HR", "CO2"]]

df_may.to_parquet("../data/002_intermediate/29-may-2026-limpio.parquet")
#df_may

```
