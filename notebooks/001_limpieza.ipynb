```python
import pandas as pd
import matplotlib.pyplot as pyplot
import numpy as np
```

```python
may = ("../data/raw/29-may-2026.csv")
torre = ("../data/raw/002_Torre2.csv")
df_may = pd.read_csv(may,
                     encoding="ISO-8859-1",
                     index_col=11,
                     parse_dates=True)

df_may

```

```python
columnas = df_may.columns
#columnas
df_may = df_may.rename (columns={
    "Temperatura":"Temp",
    "Bulbo húmedo": "BH",
    "Punto de condensación": "PC",
    "% HR": "HR",
    "Sello de tiempo": "Time" 
    })


df_may["Temp"] = df_may["Temp"].replace("°C", "", regex=True)
df_may["BH"] = df_may["BH"].replace("°C", "", regex=True)
df_may["PC"] = df_may["PC"].replace("°C", "", regex=True)
df_may["HR"] = df_may["HR"].replace("%", "", regex=True)

df_may["Temp"] = pd.to_numeric(df_may["Temp"])
df_may["BH"] = pd.to_numeric(df_may["BH"])
df_may["PC"] = pd.to_numeric(df_may["PC"])
df_may["HR"] = pd.to_numeric(df_may["HR"])

df_may = df_may.loc[:,["Lectura", "Temp", "BH", "PC", "HR"]]

df_may = df_may.rename_axis("Time")

#df_may.to_parquet("../data/processed/29-may-2026-limpio.parquet")
#df_may

```
